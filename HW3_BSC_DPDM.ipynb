### Frequent Patterns

#### Europe

```python
EU_data = mk_df[mk_df['Country']==""]
for i in range(len(EU)):
  x = mk_df[mk_df['Country']==EU[i]]
  EU_data=pd.concat([x,EU_data])
```

```python
len(set(EU_data['Country']))
```

```python
EU_data
```

```python
transactionsEU = EU_data.groupby('CustomerID')['StockCode'].apply(list)
```

```python
from operator import attrgetter

min_sup = 0.075

# สร้างรายการ Frequent Itemsets ด้วย Apriori
frequentItemsetsEU = list(apriori(transactionsEU, min_support=min_sup))

# วนลูปตรวจสอบเฉพาะ Itemsets ที่มีมากกว่า 1 รายการ และเรียงตาม confidence
sorted_frequentItemsets = sorted(
    (itemset for itemset in frequentItemsetsEU if len(itemset.items) > 1),
    key=lambda x: x.ordered_statistics[0].confidence,  # เรียงตาม confidence
    reverse=True  # เรียงจากมากไปน้อย
)

# แสดงผลลัพธ์ที่เรียงแล้ว
for itemset in sorted_frequentItemsets:
    print("yes")
    print(itemset)
```

```python
def finditem(x,y) :
  L=set(mk_df[(mk_df['StockCode']==x)|(mk_df['StockCode']==y)]['Description'])
  return L
```

```python
def finditems(x, y='', z='', w='') :
  I =set(mk_df[(mk_df['StockCode']==x)|(mk_df['StockCode']==y)|(mk_df['StockCode']==z)|(mk_df['StockCode']==w)]['Description'])
  return I
```
