```python
# แปลงค่า Target `Thal_Group` เป็นตัวเลข
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
train_data["Thal_Group"] = le.fit_transform(train_data["Thal_Group"])
```

```python
train_data['Thal_Group'].value_counts()
```

```python
# กำหนด X (features) และ y (target)
X_train = train_data.drop(columns=['id number', 'Thal_Group'])  # ลบคอลัมน์ที่ไม่เกี่ยวข้องออก
y_train = train_data['Thal_Group']  # เลือกคอลัมน์ Thal_Group เป็น target
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier

# กำหนดโมเดลที่ต้องการเปรียบเทียบ (classification models)
models = {
    'RandomForest': RandomForestClassifier(random_state=42),
    'LogisticRegression': LogisticRegression(max_iter=1000, random_state=42),
    'SVM': SVC(random_state=42),
    'KNeighbors': KNeighborsClassifier()
}
```

```python
# กำหนดพารามิเตอร์ที่ต้องการทดสอบใน GridSearchCV
param_grid = {
    'RandomForest': {
        'n_estimators': [50, 100, 200],
        'max_depth': [None, 10, 20, 30],
        'min_samples_split': [2, 5, 10],
        'min_samples_leaf': [1, 2, 4]
    },
    'LogisticRegression': {
        'C': [0.1, 1, 10],
        'solver': ['lbfgs', 'liblinear']
    },
    'SVM': {
        'C': [0.1, 1, 10],
        'kernel': ['linear', 'rbf']
    },
    'KNeighbors': {
        'n_neighbors': [3, 5, 10],
        'weights': ['uniform', 'distance']
    }
}
```

```python
from sklearn.model_selection import GridSearchCV

best_models = {}
for model_name, model in models.items():
    print(f"\nTraining {model_name}...")
    
    # ใช้ GridSearchCV
    grid_search = GridSearchCV(estimator=model, param_grid=param_grid[model_name], cv=15, scoring='accuracy')
    grid_search.fit(X_train, y_train)
    
    # หาพารามิเตอร์ที่ดีที่สุด
    best_model = grid_search.best_estimator_
    best_models[model_name] = best_model
    
    # แสดงพารามิเตอร์ที่ดีที่สุด
    print(f"Best Parameters for {model_name}: ", grid_search.best_params_)
    print(f"Best Score for {model_name}: ", grid_search.best_score_)
    print(f"Best Estimator for {model_name}: ", best_model)
    
```

```python
best_models
```

```python
from sklearn.metrics import classification_report, accuracy_score

for model_name, best_model in best_models.items():
    # ทำนายผลกับข้อมูล train
    y_pred = best_model.predict(X_train)
    
    # แสดงผลการประเมิน
    print(f"\nClassification Report for {model_name}:\n", classification_report(y_train, y_pred))
    print(f"Accuracy for {model_name}: ", accuracy_score(y_train, y_pred))
```

```python
from sklearn.model_selection import cross_val_score

for model_name, best_model in best_models.items():
    # Cross-validation scores
    cv_scores = cross_val_score(best_model, X_train, y_train, cv=5, scoring='accuracy')
    
    # แสดงผลการ Cross-validation
    print(f"\nCross-validation Accuracy Scores for {model_name} (5-fold CV): ", cv_scores)
    print(f"Mean CV Accuracy for {model_name}: ", cv_scores.mean())
    print(f"Std CV Accuracy for {model_name}: ", cv_scores.std())
    
```

```python
from sklearn.model_selection import cross_val_score

for model_name, best_model in best_models.items():
    # Cross-validation scores
    cv_scores_10 = cross_val_score(best_model, X_train, y_train, cv=10, scoring='accuracy')
    
    # แสดงผลการ Cross-validation
    print(f"\nCross-validation Accuracy Scores for {model_name} (10-fold CV): ", cv_scores_10)
    print(f"Mean CV Accuracy for {model_name}: ", cv_scores_10.mean())
    print(f"Std CV Accuracy for {model_name}: ", cv_scores_10.std())
    
```

```python
from sklearn.model_selection import cross_val_score

for model_name, best_model in best_models.items():
    # Cross-validation scores
    cv_scores_15 = cross_val_score(best_model, X_train, y_train, cv=15, scoring='accuracy')
    
    # แสดงผลการ Cross-validation
    print(f"\nCross-validation Accuracy Scores for {model_name} (15-fold CV): ", cv_scores_15)
    print(f"Mean CV Accuracy for {model_name}: ", cv_scores_15.mean())
    print(f"Std CV Accuracy for {model_name}: ", cv_scores_15.std())
    
```

```python
from sklearn.model_selection import cross_val_score

for model_name, best_model in best_models.items():
    # Cross-validation scores using F1-weighted
    cv_scores_f1_weighted = cross_val_score(best_model, X_train, y_train, cv=15, scoring='f1_weighted')
    
    # แสดงผลการ Cross-validation
    print(f"\nCross-validation F1 Weighted Scores for {model_name} (15-fold CV): ", cv_scores_f1_weighted)
    print(f"Mean F1 Weighted Score for {model_name}: ", cv_scores_f1_weighted.mean())
    print(f"Std F1 Weighted Score for {model_name}: ", cv_scores_f1_weighted.std())
```

```python
import pandas as pd
from sklearn.model_selection import cross_val_score

# สร้าง list สำหรับเก็บผลลัพธ์
results = []

# ทำการคำนวณ Cross-validation และเก็บผลลัพธ์ใน list
for model_name, best_model in best_models.items():
    # Cross-validation scores using Accuracy
    cv_scores_accuracy = cross_val_score(best_model, X_train, y_train, cv=15, scoring='accuracy')
    
    # Cross-validation scores using F1-weighted
    cv_scores_f1_weighted = cross_val_score(best_model, X_train, y_train, cv=15, scoring='f1_weighted')
    
    # เก็บผลลัพธ์ลงใน list
    results.append({
        'Model Name': model_name,
        'Accuracy Mean': cv_scores_accuracy.mean(),
        'Accuracy Std': cv_scores_accuracy.std(),
        'F1 Weighted Mean': cv_scores_f1_weighted.mean(),
        'F1 Weighted Std': cv_scores_f1_weighted.std()
    })

# สร้าง DataFrame จาก list ผลลัพธ์
results_df = pd.DataFrame(results)

# แสดงผลลัพธ์
print(results_df)
```
