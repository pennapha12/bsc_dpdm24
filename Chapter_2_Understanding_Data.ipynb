```python
# prompt: read json

import json
import pandas as pd

# Replace 'your_file.json' with the actual path to your JSON file in Google Drive
file_path = '/content/drive/MyDrive/for_work/Year3_2/DataPrepareration_and_DataMining/memo_2298.json'

try:
    with open(file_path, 'r') as f:
        data = json.load(f)

    # Check if the JSON data is a list of dictionaries
    if isinstance(data, list) and all(isinstance(item, dict) for item in data):
        df = pd.DataFrame(data)
        print(df.head())
    elif isinstance(data, dict):
        df = pd.DataFrame([data])
        print(df.head())
    else:
        print("JSON file structure not recognized. It should be a list of dictionaries or a single dictionary")

except FileNotFoundError:
    print(f"File not found at {file_path}")
except json.JSONDecodeError:
    print(f"Invalid JSON format in {file_path}")
except Exception as e:
    print(f"An error occurred: {e}")
```

# Quiz 3
print actor name and action process for first action history

```python
import json

try:
    # โหลดไฟล์ JSON
    with open('/content/drive/MyDrive/for_work/Year3_2/DataPrepareration_and_DataMining/memo_2298.json', 'r') as file:
        data = json.load(file)

    # แสดงข้อมูลทั้งหมด
    print(json.dumps(data, indent=4, ensure_ascii=False))  # ใช้ ensure_ascii=False เพื่อให้รองรับภาษาไทย

except FileNotFoundError:
    print("ไม่พบไฟล์")
except Exception as e:
    print(f"เกิดข้อผิดพลาด: {e}")
```

```python
# ดึงข้อมูลรายการแรกใน ActionHistory
first_action = data.get("ActionHistory", [])[0]  # ดึงรายการแรกของ ActionHistory

# ตรวจสอบโครงสร้างข้อมูล
if first_action:
    actor_name = first_action.get("ActorName", "ไม่มีข้อมูลชื่อผู้กระทำ")
    action_process = first_action.get("ActionProcess", "ไม่มีข้อมูลกระบวนการ")
    print(f"Actor Name: {actor_name}")
    print(f"Action Process: {action_process}")
else:
    print("ไม่มีข้อมูลใน ActionHistory")
```
